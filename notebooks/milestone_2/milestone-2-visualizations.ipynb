```python
import pandas as pd
from ggplot import *
```

```python
df = pd.read_json('themoviedb-5k-sample-with_genres-details-english.json')
```

```python
len(df)
```

```python
df.columns
```

```python
df.tail(3)
```

```python
df['cast_count'] = df.apply (lambda row: len(row['cast']),axis=1)
```

```python
df['crew_count'] = df.apply (lambda row: len(row['crew']),axis=1)
```

```python
df['genre_count'] = df.apply (lambda row: len(row['genre_ids']),axis=1)
```

```python
df['first_genre'] = df.apply (lambda row: row['genres'][0]['name'],axis=1)
```

```python
ggplot(df, aes(x='genre_count')) +\
    geom_histogram(position = 'stack', stat = 'bin') + \
    theme(title=element_text(size=20, text = 'Genre Counts per Movies'),
         axis_title_x=element_text(text='Genre Counts', size=20))
```

```python
ggplot(df, aes(x='crew_count')) +\
    geom_histogram(position = 'stack', stat = 'bin') + facet_wrap("first_genre",scales='fixed') + \
    theme(title=element_text(size=20, text = 'Crew count per genre'),
         axis_title_x=element_text(text='Genre', size=20))
```

```python
ggplot(df, aes(x='cast_count')) +\
    geom_histogram(position = 'stack', stat = 'bin') + facet_wrap("first_genre",scales='fixed') + \
    theme(title=element_text(size=20, text = 'Cast count per Genre'),
         axis_title_x=element_text(text='Genre', size=20))
```

```python
ggplot(df, aes(x='runtime')) +\
    geom_histogram(position = 'stack', stat = 'bin') + facet_wrap("first_genre",scales='fixed') + \
    theme(title=element_text(size=20, text = 'Runtime per genre'),
         axis_title_x=element_text(text='Genre', size=20))
```

```python
ggplot(df, aes(x='revenue')) +\
    geom_histogram(position = 'stack', stat = 'bin') + facet_wrap("first_genre",scales='fixed') + \
    theme(title=element_text(size=20, text = 'Runtime per genre'),
         axis_title_x=element_text(text='Genre', size=20))
```

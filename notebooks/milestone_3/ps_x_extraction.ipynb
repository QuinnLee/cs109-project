# Extracting names of cast and director/producer

```python
# Libraries
import pandas as pd
```

## Reading the data

```python
genres = pd.read_json('data/sample_expanded_genres.json')
print(genres.shape)
genres.head()
```

```python
# Unique identifier
len(genres["id"].unique())

# Empty data frame to store results
results = pd.DataFrame({"id": genres["id"]})
results["director"] = "MISS"
results["producer"] = "MISS"
results["cast1"] = "MISS"
results["cast2"] = "MISS"
results["cast3"] = "MISS"

# Looping to get names
for i in genres.index:
    if i%1000 == 0:
        print(i)
    crew = pd.DataFrame(genres.ix[i, "crew"])
    if crew.shape[0] >= 1:
        try:
            results.ix[i, "director"] = crew.ix[crew["job"] == "Director", "name"].iloc[0]
            results.ix[i, "producer"] = crew.ix[crew["job"] == "Producer", "name"].iloc[0]
            # need to correct for multiple producers
        except:
            pass
    cast = pd.DataFrame(genres.ix[i, "cast"])
    if cast.shape[0] >= 1:
        try:
            results.ix[i, "cast1"] = cast.ix[cast["order"] == 1, "name"].iloc[0]
            results.ix[i, "cast2"] = cast.ix[cast["order"] == 2, "name"].iloc[0]
            results.ix[i, "cast3"] = cast.ix[cast["order"] == 3, "name"].iloc[0]
        except:
            pass

        
# Getting to the movie level
results_unique = results.drop_duplicates(subset= ["id"], keep= "first")        
results_unique 
```
